# Popularity based recommandation

```python
# Grouping the DataFrame 'df' by 'product_name' and counting the number of occurrences of 'product_rating'.
rating_product = df.groupby('product_name').count()['product_rating'].reset_index()
rating_product.rename(columns={"product_rating":"num_rating"},inplace = True)
rating_product
```

```python
# Grouping the DataFrame 'df' by 'product_name' and calculating the average of 'product_rating' for each product.

avg_rating_product = df.groupby('product_name').mean()['product_rating'].reset_index()
avg_rating_product.rename(columns={"product_rating":"avg_num_rating"},inplace = True)
avg_rating_product
```

```python
popularity_df = rating_product.merge(avg_rating_product,on = "product_name")
popularity_df
```

```python
popularity_df.describe()
```

```python
# Filtering and selecting the top 50 popular items from a DataFrame.
# We select items with at least 3 ratings, sort them by average rating in descending order
popular_df=popularity_df[popularity_df['num_rating']>=3].sort_values('avg_num_rating',ascending=False).head(50)
```

# Collaborative Recommandation System

```python
# We group the DataFrame 'df' by 'user_id', count the occurrences of 'product_rating',
# and then select user IDs where the count is greater than 7.

x = df.groupby('user_id').count()['product_rating']>7
y = x[x].index
```

```python
filter_rating = df[df['user_id'].isin(y)]
```

```python
z = filter_rating.groupby('product_name').count()['product_rating']
famous_product = z[z].index
```

```python
final_rating=filter_rating[filter_rating['product_name'].isin(famous_product)]
```

```python
len(final_rating)
```

```python
# Creating a pivot table 'pt' from the 'final_rating' DataFrame.
# We pivot the data to have 'product_name' as rows, 'user_id' as columns,
# and 'product_rating' as the values.
pt = final_rating.pivot_table(index = 'product_name',columns='user_id',values='product_rating')
```

```python
pt.shape
```

```python
pt.fillna(0,inplace = True)
```

```python
pt
```

```python
# Cosine similarity is a metric used to measure the similarity between two vectors
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score.shape
```

```python
# Recommendation function to find similar products based on user ratings and product metadata.

def recommand(product):
    # fetch index
    index = np.where(pt.index == product)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]
    
    data = []
    for i in similar_items:
        item = []
        temp = df[df['product_name']==pt.index[i[0]]]
        item.extend(list(temp.drop_duplicates('product_name')['product_name'].values))
        item.extend(list(temp.drop_duplicates('product_name')['image_url'].values))
        item.extend(list(temp.drop_duplicates('product_name')['retail_price'].values))
        item.extend(list(temp.drop_duplicates('product_name')['discounted_price'].values))
        item.extend(list(temp.drop_duplicates('product_name')['description'].values))
        
        
        data.append(item) 
    return data
```

```python
#Calling of recommend function 

recommand('3a AUTOCARE Car Mat Honda City')
```
